# Question 8
## a)
###### <font size="3" color="red"> use erosion and dilation several time along side with bitwise or to finde the sckeleton of the image</font>

```python
img10 = cv2.imread("HotBallon.png" , cv2.IMREAD_GRAYSCALE)
img9 = cv2.imread("Dog.png" , cv2.IMREAD_GRAYSCALE)
img8 = cv2.imread("Girl.png" , cv2.IMREAD_GRAYSCALE)


imgArray = [img10 , img9 , img8]
namesArray=["ballon" , "dog" , "girl" ]
i=0
for img in imgArray:
    size = np.size(img)
    skel = np.zeros(img.shape ,np.uint8)

    ret,img = cv2.threshold(img,127,255,0)
    img = 255-img
    #plt.title('aaaaaaaaaaaaaaaaaaa'),plt.imshow(img,cmap='gray')

    element = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
    done = False
    while (not done):
        eroded = cv2.erode(img,element)
        temp = cv2.dilate(eroded,element)
        temp = cv2.subtract(img , temp)
        skel = cv2.bitwise_or(skel ,temp)
        img = eroded.copy()
        zeros = size - cv2.countNonZero(img)
        if zeros == size:
            done = True
    imgArray[i]=skel   
    plt.figure(figsize=(10,10))
    plt.title(namesArray[i]),plt.imshow(skel , cmap = 'gray')
    plt.show()
    i=i+1
```
